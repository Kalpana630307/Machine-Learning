```python
import numpy as np
import pandas as pd

# Load the dataset (Ensure to replace path with your actual path)
path = 'path_to_your_file'  # Replace this with the actual path to your 'enjoysport.csv' file
data = pd.read_csv('/content/plant_growth_data.csv')
# Extracting the features (concepts) and the target values
concepts = np.array(data.iloc[:, 0:-1])  # All columns except the last column
print("\nInstances are:\n", concepts)

target = np.array(data.iloc[:, -1])  # The last column is the target value
print("\nTarget Values are: ", target)

# Candidate Elimination Learning Algorithm
def learn(concepts, target):
    # Initialize specific_h as the first example
    specific_h = concepts[0].copy()
    print("\nInitialization of specific_h and general_h")
    print("\nSpecific Boundary: ", specific_h)

    # Initialize general_h to be the most general hypothesis
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
    print("\nGeneric Boundary: ", general_h)

    # Loop over all the instances in the dataset
    for i, h in enumerate(concepts):
        print("\nInstance", i + 1, "is", h)

        if target[i] == "yes":  # Positive instance
            print("Instance is Positive")
            # Update the specific and general boundaries for positive instances
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'  # Make the specific hypothesis more general
                    general_h[x][x] = '?'  # Update the general boundary to match
        elif target[i] == "no":  # Negative instance
            print("Instance is Negative")
            # Update the general boundary for negative instances
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]  # Make the general boundary more specific
                else:
                    general_h[x][x] = '?'  # Keep '?' if there's no match
        # Print the boundaries after processing each instance
        print("Specific Boundary after", i + 1, "Instance is", specific_h)
        print("Generic Boundary after", i + 1, "Instance is", general_h)

    # Remove irrelevant general hypotheses (those with all '?'s)
    print("\nCleaning the generic boundary...")
    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?', '?']]
    for i in indices:
        general_h.remove(['?', '?', '?', '?', '?', '?'])

    return specific_h, general_h

# Call the learn function to generate the final specific and general boundaries
s_final, g_final = learn(concepts, target)

# Output the final specific and general hypotheses
print("\nFinal Specific_h: ", s_final, sep="\n")
print("\nFinal General_h: ", g_final, sep="\n")
```
